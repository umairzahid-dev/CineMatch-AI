# CineMatch AI - Movie Recommendation System

## Project Overview

CineMatch AI is a content-based movie recommendation system that suggests similar movies based on user-selected movies.

The main goal of this project is to build an end-to-end machine learning application that combines data preprocessing, natural language processing, similarity-based recommendation techniques, and a user-friendly web interface.

In this project, movie metadata such as genres, keywords, cast, crew, and descriptions are analyzed to understand the characteristics of each movie. The system uses TF-IDF vectorization and cosine similarity to find and recommend movies with similar content.

## Feature Engineering

In this section, we will transform raw movie information into meaningful features for our recommendation system.

Different attributes such as genres, keywords, cast, crew, and overview will be combined to create a single text-based feature that represents each movie.

```python
movies.head()
```

```python
movies['genres'][0]
```

```python
import ast
```

```python
def convert(obj):
    List = []
    for i in ast.literal_eval(obj):
        List.append(i['name'])
    return List
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['genres'].head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies['genres'].head()
```

**For Top 3 cast**

```python
def convert_cast(obj):
    List = []
    counter = 0

    for i in ast.literal_eval(obj):
        if counter < 3:
            List.append(i['name'])
            counter += 1
        else:
            break

    return List
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
movies['cast'].head()
```

**For Director**

```python
def fetch_director(obj):
    List = []

    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            List.append(i['name'])
            break

    return List
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies['crew'].head()
```

```python
movies.rename(columns={'crew':'director'}, inplace=True)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['overview'].isnull().sum()
```

```python
movies.dropna(subset=['overview'], inplace=True)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

## Creating Combined Features

To build a content-based recommendation system, we combine different movie attributes into a single text feature.

The combined feature includes movie overview, genres, keywords, cast, and director information.

This combined text will be used by the TF-IDF algorithm to understand similarities between movies.

```python
movies['tags'] = (
    movies['overview'] +
    movies['genres'] +
    movies['keywords'] +
    movies['cast'] +
    movies['director']
)
```

```python
movies[['tags']].head()
```

```python
movies['tags'] = movies['tags'].apply(lambda x: " ".join(x))
```

```python
movies['tags'] = movies['tags'].apply(lambda x:x.lower())
```

```python
def recommend(movie):

    # Find selected movie index
    index = movies[movies['title'] == movie].index[0]

    # Get similarity scores of selected movie
    distances = similarity[index]

    # Sort movies based on similarity score
    movies_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x:x[1]
    )[1:6]


    recommendations = []

    for i in movies_list:
        recommendations.append({
            "title": movies.iloc[i[0]].title,
            "similarity_score": round(i[1], 2)
        })

    return recommendations
```
